```python
# create histogram and comment

import matplotlib.pyplot as plt

ratings = df.Rating.values

plt.figure(figsize = (16,5))
plt.hist(ratings, bins = 4, color = 'red')
plt.xlabel('Ratings')

# looks like most of chocolates have ratings between 3 and 4
```

# Visualization

```python
# Convert and visualize (17:00)

df["CocoaPercent"] = df["CocoaPercent"].str.replace('%',"").astype(float)
df
```

```python
ratings = df.Rating.values
percent = df.CocoaPercent.values

plt.figure(figsize = (16,5))
plt.scatter(percent, ratings, color = 'red', alpha = .1)
plt.xlabel('Cocoa Percent')
```
